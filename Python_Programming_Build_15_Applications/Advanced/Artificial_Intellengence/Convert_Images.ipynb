```python
# import 'opencv'
import cv2

# Read an image
img = cv2.imread('sample.jpg')

# Show the image
cv2.imshow('Image', img)

# The image will be open in an GUI application, and has to have this 
# 'waitKey' as well, or the application will not respond
cv2.waitKey(0)
```

```python
# The same can be done (to view it in notebook) with matplotlib
from matplotlib.pyplot import imshow

img = cv2.imread('sample.jpg')
imshow(img)

# The color of the image is different though then was previously viewed with 'cv2.imshow()'
# RGB - Red, Green, Blue    (used by every software)
# BGR - Blue, Green, Red    (used by 'opencv')
```

```python
# Change the color space
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
imshow(img_rgb)
```

```python
# Create a black and white image instead
img_bw = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
imshow(img_bw)
```

```python
cv2.imshow('Black & White', img_bw)
cv2.waitKey(0)
```
