```python
import pandas

dt = pandas.read_csv('csv_data.csv')
dt.fillna(0, inplace = True)
```

```python
from matplotlib import pyplot

# Make a plot chart
pyplot.plot(dt['id'], dt['sales'])
```

```python
# Make a bar chart
pyplot.bar(dt['id'], dt['sales'])
```

```python
# Make a scatter chart, with the data points inlarged, and with different color
pyplot.scatter(dt['id'], dt['sales'], linewidth = 60, color = 'orange')
```

```python
# Make a bar chart horizontally aligned
pyplot.barh(dt['id'], dt['sales'])
```

```python
# See the average (mean) sale over the sales per ID
pyplot.barh(dt['id'], dt['sales'])
pyplot.barh(dt['id'], dt['sales'].mean())
```

```python
# Or see it the other way around (sales over the average sale number)
pyplot.barh(dt['id'], dt['sales'].mean())
pyplot.barh(dt['id'], dt['sales'])
```

```python
# Add a legend to the graph, indicating the colors (labels) of the chart 
pyplot.barh(dt['id'], dt['sales'].mean(), label = 'Avg. Sales')
pyplot.barh(dt['id'], dt['sales'], label = 'Act. Sales')
pyplot.legend()
```

```python
# Make a pie-chart of the data of sales, with their names accordingly
pyplot.pie(dt['sales'], labels = dt['name'])
```

```python
# Make the pie-chart larger
pyplot.pie(dt['sales'], labels = dt['name'], radius = 3)
# pyplot.legend()
```
