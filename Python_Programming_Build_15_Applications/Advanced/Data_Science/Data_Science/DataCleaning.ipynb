# Adding manually a column header (and remove previous header)

# Remove / Replace values within data

```python
# Replace null values with 0:
dt.fillna(0)

# Replace null values with Null:
dt.fillna('Null')

# Inherite the values above the missing value:
dt.pad()

# Inherite the values below the missing value:
dt.bfill()

# Remove all rowes in data with null values (NaN):
dt.dropna()

# Remove a specific row in the data (here, rows 1 and 5):
dt.drop([1,5])

# Delete specified column:
del dt['Price']
dt

# Drop specific columns, and store in another dataframe (will not have lasting effect on original data) 
# NB: Can not be run simultaniously with previous commands:
#    new_data = dt.drop(columns = ['Sales', 'Price'])
```
