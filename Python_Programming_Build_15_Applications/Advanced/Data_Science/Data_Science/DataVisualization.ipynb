```python
import pandas
dt = pandas.read_csv('csv_data.csv')

# Drop the lines in data that have NaN-values and store with inplace-method
#     dt.dropna(inplace = True)

# Fill NaN-values with 0 instead within the data
dt.fillna(0, inplace = True)

# Drop the ID column within data, as we don't need that for visualization here
#     dt.drop(columns = ['id'], inplace = True)
```

```python
# Plot the data in a diagram
dt.plot()
```

```python
# Plot only the Sales column in a diagram
dt['sales'].plot()
```

```python
# Plot the Sales column in a bar diagram
dt['sales'].plot.bar()
```

```python
# Plot the Sales column in a bar diagram, horizontally
dt['sales'].plot.barh()
```

```python
# Plot and make a comparison between data-fields as a bar diagram
dt.plot.bar()
```

```python
# Have the same diagram, but stacked values of both columns
dt.plot.bar(stacked = True)
```

```python
# Have the same diagram, but stacked values of both columns, horizontally
dt.plot.barh(stacked = True)
```

```python
# Plot data as area plot
dt.plot.area()
```

```python
# Plot data as area plot, but not stacked
dt.plot.area(stacked = False)
```

```python
# Make a histogram
dt.plot.hist()
```

```python
# Get comparison with the total sales, actual sales, and average sales, in a bar diagram (stacked)
dt_analytics = pandas.DataFrame({'Actual Sales': dt['sales'],
                                 'Total Sales': dt['sales'].sum(),
                                 'Average Sales': dt['sales'].mean() })
dt_analytics.plot.bar(stacked = True)
```

```python
# The same chart as above, but not stacked...
dt_analytics.plot.bar()
```
