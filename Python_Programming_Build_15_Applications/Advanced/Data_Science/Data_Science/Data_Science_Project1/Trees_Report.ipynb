```python
import pandas as pd

# Import data, with given headers (names)
data = pd.read_csv('trees.csv', names = ['I', 'Girth', 'Height', 'Volume'])

# Delete the index column
del data['I']

# Remove the first row within data, as it contains the old headers
data.drop(0, inplace = True)

# Change measurement of of columns to float, 
# so we can operate mathematical function with these data values
data['Girth'] = data['Girth'].astype(float)
data['Height'] = data['Height'].astype(float)
data['Volume'] = data['Volume'].astype(float)

# 'Girth' is in inches, and shall be changed to metres (1 inch = 0.0254 metres)
data['Girth'] *= 0.0254

# 'Height' is in foot, and shall be changed to metres (1 foot = 0.305 metres)
data['Height'] *= 0.305

# 'Volume' is in cubic foot, ans shall be changes to metres (1 cubic foot = 0.0283 cubic metres)
data['Volume'] *= 0.0283
```

```python
# Find the maximum values
max_girth = (data['Girth'].round(2)).max()
max_height = (data['Height'].round(2)).max()
max_volume = (data['Volume'].round(2)).max()

# Find the minimum values
min_girth = (data['Girth'].round(2)).min()
min_height = (data['Height'].round(2)).min()
min_volume = (data['Volume'].round(2)).min()

# FInd the mean values
avg_girth = round(data['Girth'].mean(), 2)
avg_height = round(data['Height'].mean(), 2)
avg_volume = round(data['Volume'].mean(), 2)

# Find the median values
median_girth = (data['Girth'].round(2)).median()
median_height = (data['Height'].round(2)).median()
median_volume = (data['Volume'].round(2)).median()

# Find the mode values (most frequen value)
mode_girth = (data['Girth'].round(2)).mode()
mode_height = (data['Height'].round(2)).mode()
mode_volume = (data['Volume'].round(2)).mode()
```
